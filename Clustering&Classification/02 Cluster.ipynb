```python
import pandas as pd
import numpy as np
from joblib import load, dump
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv("data_new.csv")
```

```python
data
```

```python
X = data.values
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
kmeans = KMeans(n_clusters = 3)
```

```python
y_pred = kmeans.fit_predict(X)
CH = metrics.calinski_harabasz_score(X, y_pred)
silhouette = metrics.silhouette_score(X, y_pred)
print("n_cluster = %d, calinski harabasz score = %f, silhouette score = %f"%(3, CH, silhouette))
```

```python
data['Cluster'] = y_pred
```

```python
print('{}\t{}\t{}\t\t{}\t{}\t\t{}'.format('Cluster', 'Temperature', 'Pressure', 'Humidity', 'Smoke', 'Flame'))
print('{}'.format('-' * 100))

for Cluster in data.Cluster.unique():
    print('{:d}\t{:.2f} ({:.2f})\t{:.2f} ({:.2f})\t{:.2f} ({:.2f})\t{:.2f} ({:.2f})\t{:.2f} ({:.2f})'.format(
        Cluster
        , data[data.Cluster==Cluster].Temperature.mean()
        , data[data.Cluster==Cluster].Temperature.std()
        , data[data.Cluster==Cluster].Pressure.mean()
        , data[data.Cluster==Cluster].Pressure.std()
        , data[data.Cluster==Cluster].Humidity.mean()
        , data[data.Cluster==Cluster].Humidity.std()
        , data[data.Cluster==Cluster].Smoke.mean()
        , data[data.Cluster==Cluster].Smoke.std()
        , data[data.Cluster==Cluster].Flame.mean()
        , data[data.Cluster==Cluster].Flame.std()
    ))

print('\n{}\n'.format('=' * 100))
```
