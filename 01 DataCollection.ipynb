```python
import pandas as pd
import numpy as np
from random import random
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv("data.txt", sep = ";", header = None,
                 names = ['Temperature', 'Pressure', 'Humidity', 'Smoke', 'Flame'])
```

```python
for col in data:
    data[col] = data[col].str.split(':').str[1].astype('float64')

data = data.where(data != -1)
```

```python
diff = data.Pressure[data.Pressure > 100000].mean() - data.Pressure[data.Pressure < 100000].mean()
data.Pressure[data.Pressure < 100000] = data.Pressure[data.Pressure < 100000] + int(diff)
```
